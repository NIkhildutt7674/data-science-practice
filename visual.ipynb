```python
import os
os.chdir(r"C:\Users\my pc\Desktop\data_analysis")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Toyota.csv",index_col=0,na_values=['??','????'])
print(data) 
```

```python
data.dropna(axis=0,inplace=True)
data.shape
```

```python
plt.scatter(data['Age'],data['Price'],c='blue')
plt.title("Scatter plot between Age and Price")
plt.xlabel("Age")
plt.ylabel("Price")
plt.show()
```

```python
plt.hist(data['KM'],color='green',edgecolor='white',bins=5)
plt.title('Histogram of Km')
plt.xlabel("km")
plt.ylabel("Frequency")
```

```python
count=[979,120,12]
FuelType=('Petrol','Diesel','CNG')
index=np.arange(len(count))
#plt.bar(index,count,color=['red','green','blue'])
plt.bar(FuelType,count,color=['red','green','blue'])
plt.title('bar plot of FuelType')
plt.xlabel("Fueltypes")
plt.ylabel("Frequency")
#plt.xticks(index,FuelType)
```

```python
import seaborn as sns
```

```python
sns.regplot(x=data['Age'],y=data['Price'],marker='*',fit_reg=False)
sns.lmplot(x='Age',y="Price",data=data,fit_reg=False,hue='FuelType',legend=True,palette='Set1')
```

```python
sns.distplot(data['Age'],kde=False,bins=5)
```

```python
#sns.countplot(x=data['FuelType])
sns.countplot(x='FuelType',data=data)
plt.show()
sns.countplot(x='FuelType',data=data,hue="Automatic")
```

```python
sns.boxplot(y=data["Price"])
plt.show()
sns.boxplot(x=data['FuelType'],y=data['Price'])
plt.show()
sns.boxplot(x=data['FuelType'],y=data['Price'],hue=data['Automatic'])
```
