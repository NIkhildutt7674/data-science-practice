Student Marks Analytics System
🎯 Goal:
Work only with NumPy arrays.
import numpy as np
marks = np.array([
    [85, 78, 92],
    [70, 88, 90],
    [60, 75, 85],
    [95, 92, 98]
])
Tasks:
Calculate student-wise average
Subject-wise average
Highest scorer
Rank students using total marks
Normalize marks between 0 and 1
Add bonus marks only if average < 80

```python
import numpy as np
marks = np.array([
    [85, 78, 92],
    [70, 88, 90],
    [60, 75, 85],
    [95, 92, 98]
])
stu_avg = np.mean(marks, axis=1)
print(stu_avg)
sub_avg=np.mean(marks,axis=0)
print(sub_avg)
print(max(stu_avg))
```

```python
stu_sum=np.sum(marks,axis=1)
top_student = np.argmax(stu_sum)
print(top_student+1," is the highest scorer")
```

```python
rank=np.argsort(stu_sum)[::-1]
print(rank)
```

```python
min_val = np.min(marks)
max_val = np.max(marks)
normalized = (marks - min_val) / (max_val - min_val)
print(normalized)
```

Sales Data Analyzer
sales = np.array([
    [200, 250, 300],
    [150, 180, 210],
    [400, 420, 390],
    [100, 130, 160]
])

Each row = Store
Each column = Month

Tasks:
Total sales per store
Best performing month
Store with highest growth
Percentage increase month-wise
Standard deviation per store

```python
sales = np.array([
    [200, 250, 300],
    [150, 180, 210],
    [400, 420, 390],
    [100, 130, 160]
])
t_s=np.sum(sales,axis=1)
print(t_s)
```

```python
t_m=np.sum(sales,axis=0)
b_m=np.argmax(t_m)
print(t_m)
print("best performing month:",b_m+1)
```

```python
growth=sales[:,-1]-sales[:,0]
print(growth)
h_g=np.argmax(growth)
print(h_g+1)
```

```python
perc_1_2 = ((sales[:,1] - sales[:,0]) / sales[:,0]) * 100
print(perc_1_2)
perc_2_3 = ((sales[:,2] - sales[:,1]) / sales[:,1]) * 100
print(perc_2_3)
```

```python
std_dev = np.std(sales, axis=1)
print(std_dev)
```

Data Cleaning Simulator
data = np.array([10, 20, -5, 30, -2, 50, 1000])
Tasks:
Remove negative values
Replace outliers (>100) with mean
Standardize data (Z-score normalization)
Min-Max scaling

```python
data = np.array([10, 20, -5, 30, -2, 50, 1000])
data=data[data>=0]
print(data)
```

```python
mean=np.mean(data)
data[(data>100)]=mean
print(data)
```

```python
std=np.std(data)
stand=(data-mean)/std
print(stand)
```

```python
min=np.min(data)
max=np.max(data)
minmax=(data-min)/(max-min)
print(minmax)
```
