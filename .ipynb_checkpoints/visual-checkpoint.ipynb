```python
data=pd.read_csv("Toyota.csv",index_col=0,na_values=['??','????'])
print(data)
```

```python
print(data.isnull().sum()) 
np.unique(data['Doors'])
```

```python
print(data['Doors'].head(20))
print(data['Doors'].unique())
```

```python
# Replace words with numbers
data['Doors'] = data['Doors'].replace({
    'three':3,
    'four':4,
    'five':5
})

# Convert to numeric
data['Doors'] = pd.to_numeric(data['Doors'], errors='coerce')

# Fill missing values with 4 (most common doors)
data['Doors'] = data['Doors'].fillna(4)

# Convert to integer
data['Doors'] = data['Doors'].astype(int)

print(np.unique(data['Doors']))
```
