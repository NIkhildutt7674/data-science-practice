```python
import numpy as np
```

```python
mylist=[1,2,3,4,5,6,7,8]
lis=[12,13,14,16,17,11,19,18]
arr=np.array(mylist,dtype=int)
arr5=np.array(lis)
arr2=np.array([mylist,lis])
arr3=np.array(mylist+lis)
```

```python
print(arr)
print(type(mylist))
print(type(arr))
print(mylist)
print(arr2,arr3)
```

```python
print(arr.ndim)
print(arr.shape)
print(arr2.shape)
print(arr3.shape)
```

```python
arr1=arr.reshape(4,2)
arr4=arr2.reshape(8,2)
print(arr1)
print(arr4)
```

```python
print(np.add(arr,arr5))
print(np.subtract(arr,arr5))
print(np.multiply(arr,arr5))
print(np.dot(arr,arr5))
print(np.divide(arr,arr5))
print(np.sum(arr5))
```

```python
print(np.mean(arr))
print(np.std(arr5))
print(np.median(arr2))
print(np.max(arr5))
```

```python
ran=np.random.randint(1, 101, 10)
print(ran)
np.random.seed(42)
print(np.random.rand(3))
```

```python
print(arr[arr > 4])
print(arr[arr % 2 == 0])
arr[arr < 5] = 0
print(arr)
```

Student Marks Analytics System
🎯 Goal:

Work only with NumPy arrays.

import numpy as np

marks = np.array([
    [85, 78, 92],
    [70, 88, 90],
    [60, 75, 85],
    [95, 92, 98]
])
Tasks:

Calculate student-wise average

Subject-wise average

Highest scorer

Rank students using total marks

Normalize marks between 0 and 1

Add bonus marks only if average < 80

```python
import numpy as np
marks = np.array([
    [85, 78, 92],
    [70, 88, 90],
    [60, 75, 85],
    [95, 92, 98]
])
stu_avg = np.mean(marks, axis=1)
print(stu_avg)
sub_avg=np.mean(marks,axis=0)
print(sub_avg)
k=max(stu_avg)
print(k)
```
